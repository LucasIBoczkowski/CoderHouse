Nuevamente el planteo se efectúa mediante dos hipótesis en mente.
- La primera hipótesis, **H<sub>0</sub>**, sostiene que no hay ninguna correlación entre los goles marcados y el ranking de los equipos locales.

<br><br>
Como siempre, usemos los datos para respaldarnos en una de las dos alternativas.

#### Desarrollo

Para este caso utilizamos los valores de home_team_fifa_rank y home_team_score, y graficamos el resultado con un scatter plot (es decir, un **gráfico de análisis bivariado**):

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Uso un scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_soccer_modified, x="home_team_fifa_rank", y="home_team_score")
plt.xlabel('Ranking FIFA de equipo', fontsize=14)
plt.xticks(fontsize=12)
plt.ylabel('Cantidad de goles', fontsize=14)
plt.yticks(fontsize=12)
plt.title('Relación entre ranking FIFA y cantidad de goles', fontsize=17)

# Muestro los resultados
plt.show()
```

A medida que exploramos los datos y analizamos las estadísticas, nos encontramos con un resultado sorprendente. El coeficiente de correlación revela un número asombrosamente cercano a cero (0): -0,12. Esto nos coloca al borde de la correlación nula, lo que significa que no existe una relación significativa entre el ranking FIFA de los equipos locales y la cantidad de goles que anotan en sus partidos.

Podemos realizar un análisis adicional, contabilizando la cantidad de goles por cada país, y su ranking FIFA.

```python
# Lista unica de selecciones
teams = pd.concat([df_soccer_modified['home_team'], df_soccer_modified['away_team']]).unique()

# Total de goles por pais
total_goals_by_team = {}

for country in teams:
    home_goals = df_soccer_modified[df_soccer_modified['home_team'] == country]['home_team_score'].sum()
    away_goals = df_soccer_modified[df_soccer_modified['away_team'] == country]['away_team_score'].sum()
    total_goals = home_goals + away_goals
    total_goals_by_team[country] = total_goals

# Dataframe con los goles por pais
df_goals_ranking = pd.DataFrame(total_goals_by_team.items(), columns=['Seleccionado', 'Total de Goles'])
# Agregamos el ranking
df_goals_ranking['Ranking'] = df_goals_ranking['Seleccionado'].apply(lambda x: df_soccer_modified.loc[df_soccer_modified['home_team'] == x, 'home_team_fifa_rank'].sum() +
                                                               df_soccer_modified.loc[df_soccer_modified['away_team'] == x, 'away_team_fifa_rank'].sum())
```

Ordenamos de mayor a menor para ver los seleccionados que más goles han convertido:

```python
df_goals_ranking = df_goals_ranking.sort_values(by='Total de Goles', ascending=False)
df_goals_ranking
```

Podemos apreciar que los equipos con más goles convertidos no necesariamente son los primeros en el Ranking FIFA. Veamos los goles por cada seleccionado en un mapa:

```python
# Creando un mapa con los goles por cada pais
fig = px.choropleth(
    df_goals_ranking,
    locations="Seleccionado",
    locationmode="country names",
    color="Total de Goles",
    hover_name="Seleccionado",
    color_continuous_scale="Viridis",
    projection="natural earth"
)

# Aumentar el tamaño del título
fig.update_layout(
    title_text="Goles convertidos por cada seleccionado",
    title_x=0.5,  # Ajusta la posición horizontal del título
    title_font=dict(size=17)  # Ajusta el tamaño del título
)

# Mostramos el mapa
fig.show()
```

De esta manera podemos observar que no hay una relación directa entre ranking y goles convertidos por seleccionado.

#### Insights y conclusión

Como pudimos apreciar mediante el gráfico y con el coeficiente de correlación, no hay relación directa, con lo cual **concluimos que la hipótesis nula H<sub>0</sub> es la acertada**.
<br><br>
Este hallazgo nos lleva a reflexionar sobre la complejidad inherente al fútbol y nos indica que estamos hablando de un deporte multifacético y complejo por naturaleza. Si bien el ranking FIFA puede ofrecer una medida de la fuerza relativa de un equipo, claramente no es el único factor determinante de su capacidad para marcar goles. El fútbol es un deporte multifacético, donde las tácticas, el rendimiento individual, la calidad del oponente y las circunstancias específicas del partido se entrelazan en un ballet caótico que da forma a los marcadores.
<br><br>
Así que, mientras el fútbol siga desafiando nuestras expectativas y rompiendo las reglas de las estadísticas, seguiremos maravillándonos ante la magia de este deporte. ### 6.5. Relación entre Ranking FIFA y cantidad de partidos ganados

#### Hipótesis

En el vasto campo de batalla del fútbol internacional, donde las selecciones nacionales luchan por la gloria en cada partido, existe una fascinante pregunta que se abre paso entre la multitud de hinchas apasionados: ¿existe una conexión entre la cantidad de partidos ganados por un equipo y su posición en el ranking FIFA?
Vamos a explorar este enigma y descubrir si la victoria en el campo de juego se traduce en un lugar más alto en la clasificación mundial.
<br><br>
En esta búsqueda de la verdad, nuevamente encaramos dos hipótesis que nos llevarán por un camino de datos y gráficos.
- La primera, la hipótesis nula **H<sub>0</sub>**, desafía la idea de que existe una relación entre el Ranking FIFA y los partidos ganados por una selección, con lo cual la redactamos diciendo que no hay relación entre Ranking FIFA y partidos ganados por selección.
- La segunda, la hipótesis alternativa **H<sub>1</sub>**, sugiere que a medida que una selección gana más partidos, su posición en el ranking FIFA se eleva.

Chequeemos esto con números y gráficos.

#### Desarrollo

Generemos un dataframe auxiliar para este caso en cuestión:

```python
df_soccer_modified_case_5 = df_soccer_modified
```

Calculamos el ranking de cada selección:

```python
team_ranking = pd.DataFrame()
team_ranking['team_name'] = pd.concat([df_soccer_modified_case_5['home_team'], df_soccer_modified_case_5['away_team']]).unique()
team_ranking['ranking'] = team_ranking['team_name'].apply(lambda x: df_soccer_modified_case_5.loc[df_soccer_modified_case_5['home_team'] == x, 'home_team_fifa_rank'].sum() +
                                                               df_soccer_modified_case_5.loc[df_soccer_modified_case_5['away_team'] == x, 'away_team_fifa_rank'].sum())
team_ranking
```

Calculamos la cantidad de partidos ganados por selección:

```python
matches_won_home = df_soccer_modified_case_5[df_soccer_modified_case_5['home_team_result'] == 'Win']['home_team'].value_counts()
matches_won_away = df_soccer_modified_case_5[df_soccer_modified_case_5['home_team_result'] == 'Lose']['away_team'].value_counts()
matches_won = matches_won_home.add(matches_won_away, fill_value=0)
matches_won
```

```python
team_stats['matches_won'] = team_stats['team_name'].apply(lambda x: matches_won[x] if x in matches_won else 0)
team_stats = team_stats.sort_values(by='ranking', ascending=True)
team_stats
```

El coeficiente de correlación es:

```python
# Calculamos el coeficiente de correlacion entre Ranking y partidos ganados
correlation = team_stats['ranking'].corr(team_stats['matches_won'])

# Imprimimos el coeficiente de correlacion
print("Coeficiente de correlación:", correlation)
```

Aquí también nos encontramos con un coeficiente de correlación que se aferra obstinadamente al cero (0). ¿Qué significa esto? En términos sencillos, estamos hablando de una correlación nula, una ausencia de relación aparente entre el ranking FIFA y la cantidad de partidos ganados. #### Hipótesis

El fútbol ha estado impregnado de la creencia común de que jugar en casa brinda una ventaja innegable. La historia del deporte está repleta de ejemplos en los que los equipos locales han triunfado en sus estadios, impulsados por el apoyo de su afición y una familiaridad inigualable con las condiciones locales. Sin embargo, también hemos sido testigos de momentos icónicos donde la localía no garantiza la victoria, como el legendario "Maracanazo", donde la selección de Uruguay se quedó con la victoria en el último partido de la Copa Mundial de Fútbol de 1950, frente a la selección de fútbol de Brasil, por 2-1 en el imponente estadio Maracaná de Río de Janeiro, Brasil. Este análisis busca arrojar luz sobre el enigma de la localía en el fútbol.
<br><br>
Vamos a plantear dos hipótesis que desafían las percepciones comunes sobre la localía.
- La primera, **H<sub>0</sub>**, sugiere que jugar en casa no otorga una ventaja significativa en términos de victoria.

#### 6.8.3. Análisis de partidos con tiempo extra y penales

Veamos cómo afecta la presencia de tiempo extra y penales al resultado final de un partido.

Primero, identifiquemos los partidos con tiempos extra y penales:

```python
# Partidos con tiempos extra y penales
extra_time_and_penalties = df_soccer_modified[df_soccer_modified['shoot_out'] == 'Yes']
# partidos sin tiempos extra ni penales
no_extra_time_or_penalties = df_soccer_modified[df_soccer_modified['shoot_out'] == 'No']
```

Ahora podemos calcular estadísticas descriptivas y realizar comparaciones entre ambos subconjuntos. Por ejemplo el promedio de goles marcados:

```python
avg_goals_with_extra_time_penalties = extra_time_and_penalties['home_team_score'].mean()
avg_goals_without_extra_time_penalties = no_extra_time_or_penalties['home_team_score'].mean()

print(f"Promedio de goles con tiempos extra y penales: {avg_goals_with_extra_time_penalties}")
print(f"Promedio de goles sin tiempos extra ni penales: {avg_goals_without_extra_time_penalties}")
```

O bien, podríamos ver el porcentaje de victorias, empates y derrotas, analizando la proporción respectiva en cada subconjunto.

```python
def calculate_win_draw_loss_percentage(df):
    total_matches = len(df)
    wins = len(df[df['home_team_result'] == 'Win'])
    draws = len(df[df['home_team_result'] == 'Draw'])
    losses = len(df[df['home_team_result'] == 'Lose'])

    win_percentage = (wins / total_matches) * 100
    draw_percentage = (draws / total_matches) * 100
    loss_percentage = (losses / total_matches) * 100

    return win_percentage, draw_percentage, loss_percentage

win_draw_loss_with_extra_time_penalties = calculate_win_draw_loss_percentage(extra_time_and_penalties)
win_draw_loss_without_extra_time_penalties = calculate_win_draw_loss_percentage(no_extra_time_or_penalties)

print("Porcentaje de victorias, empates y derrotas con tiempos extra y penales:")
print(f"Victorias: {win_draw_loss_with_extra_time_penalties[0]}%")
print(f"Empates: {win_draw_loss_with_extra_time_penalties[1]}%. Lógicamente esto es 0, pues cuando hay penales sí o sí un equipo se queda con la victoria.")
print(f"Derrotas: {win_draw_loss_with_extra_time_penalties[2]}%")

print("\nPorcentaje de victorias, empates y derrotas sin tiempos extra ni penales:")
print(f"Victorias: {win_draw_loss_without_extra_time_penalties[0]}%")
print(f"Empates: {win_draw_loss_without_extra_time_penalties[1]}%")
print(f"Derrotas: {win_draw_loss_without_extra_time_penalties[2]}%")
```

De esta manera, podemos notar que cuando llegamos a la tanda de penales, las probabilidades de victoria para el equipo visitante aumentan. Esto se debe a que la brecha entre ganar y perder para un equipo local durante los penales es mucho más estrecha que la diferencia que se observa cuando no se llega a esa fase.

#### 7.3.3. Estandarización de variables numéricas

Vamos a normalizar las variables numéricas de interés de acuerdo con los valores mínimos y máximos de cada variable, para lo cual definimos una función que nos permita realizar esta acción.

```python
# Funcion para normalizar de acuerdo con los valores mínimos y máximos de cada variable
def normalize(df):
    result = df.copy()

    for feature_name in df.columns:
        max_val = df[feature_name].max()
        min_val = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_val) / (max_val - min_val)

    return result
```

#### 8.1.6. Support Vector Machine (SVM)

Para el caso de SVM, vamos a comparar primero diferentes valores de hiperparámetros C y kernel para ver con cuál nos quedamos, y luego sumar el mismo análisis que ya fue mostrado con el resto de los modelos.

```python
Cs = [0.01, 0.1, 1, 10]
kernels = ["linear", "poly", "rbf"]
# osea, 12 combinaciones
df_results_prueba = pd.DataFrame()

for C in Cs:
  for kernel in kernels:
    modelo_svm = SVC (C = C, kernel = kernel, random_state = 42)
    modelo_svm.fit(X_train, y_train)

    preds_val_svm = modelo_svm.predict(X_val)

    #acc_val = accuracy_score(y_val, preds_val_svm)
    val_f1 = f1_score(y_val, preds_val_svm, average='macro', zero_division=1)

    res = pd.DataFrame({"C": [C], "Kernel": [kernel], "F1_Val": [val_f1]})
    df_results_prueba = pd.concat([df_results_prueba, res], axis = 0)

df_results_prueba
```

Se observa que los mejores valores se obtienen con un kernel lineal y un valor de C = 0.01.

Un valor bajo de C implica que tenemos un margen más amplio alrededor del hiperplano de decisión. Esto significa que el modelo tiende a ser más suavizado y generalizado, menos propenso al sobreajuste y puede tener una mejor capacidad de generalización a datos nuevos o de prueba. Pero por otro lado, también implica hay una mayor tolerancia a los errores de clasificación y más puntos de datos se pueden encontrar dentro del margen o en el lado incorrecto del hiperplano de decisión sin penalizarlos fuertemente. Y no buscamos esto, por lo cual vamos a aumentar el valor de C, usando el valor por defecto (C=1.0).

```python
# Creo el modelo SVM lineal
model = SVC(C = 1.0, kernel='linear')
# Entreno el modelo SVM con los datos de entrenamiento
model.fit(X_train, y_train)

# Predecir con la data de train
y_train_pred = model.predict(X_train)
# Predecir con la data de val
y_val_pred = model.predict(X_val)
# Entrenar el modelo con trainval
model.fit(X_trainval, y_trainval)

# Predecir con la data de test
y_test_pred = model.predict(X_test)

# Métricas:

# accuracy
train_accuracy = accuracy_score(y_train, y_train_pred)
val_accuracy = accuracy_score(y_val, y_val_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)
print("Accuracy en training:", train_accuracy)
print("Accuracy en validation:", val_accuracy)
print("Accuracy en test:", test_accuracy)

# precision
train_precision = precision_score(y_train, y_train_pred, average='macro', zero_division=1)
val_precision = precision_score(y_val, y_val_pred, average='macro', zero_division=1)
test_precision = precision_score(y_test, y_test_pred, average='macro', zero_division=1)
print("Precision en training:", train_precision)
print("Precision en validation:", val_precision)
print("Precision en test:", test_precision)

# Recall o sensibilidad
train_recall = recall_score(y_train, y_train_pred, average='macro', zero_division=1)
val_recall = recall_score(y_val, y_val_pred, average='macro', zero_division=1)
test_recall = recall_score(y_test, y_test_pred, average='macro', zero_division=1)
print("Recall en training:", train_recall)
print("Recall en validation:", val_recall)
print("Recall en test:", test_recall)

# F1-score
train_f1 = f1_score(y_train, y_train_pred, average='macro', zero_division=1)
val_f1 = f1_score(y_val, y_val_pred, average='macro', zero_division=1)
test_f1 = f1_score(y_test, y_test_pred, average='macro', zero_division=1)
print("F1-Score en training:", train_f1)
print("F1-Score en validation:", val_f1)
print("F1-Score en test:", test_f1)

# Classification report
train_report = classification_report(y_train, y_train_pred, zero_division=1)
print("\nClassification report en training:\n", train_report)
val_report = classification_report(y_val, y_val_pred, zero_division=1)
print("\nClassification report en validation:\n", val_report)
test_report = classification_report(y_test, y_test_pred, zero_division=1)
print("\nClassification report en test:\n", test_report)

# confusion matrix
train_cm = confusion_matrix(y_train, y_train_pred)
print("\nConfusion Matrix en training:\n")
print(train_cm)
val_cm = confusion_matrix(y_val, y_val_pred)
print("\nConfusion Matrix en validation:\n")
print(val_cm)
test_cm = confusion_matrix(y_test, y_test_pred)
print("\nConfusion Matrix en test:\n")
print(test_cm)

df_f1_results = pd.DataFrame({"Modelo": "SVM Lineal", "F1 en training": [train_f1], "F1 en validation": [val_f1], "F1 en test": [test_f1]})
df_f1_metrics = pd.concat([df_f1_metrics, df_f1_results], axis = 0).reset_index(drop=True)
```
